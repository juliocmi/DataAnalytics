**Descripción de los datos**

Megaline redondea los segundos a minutos y los megabytes a gigabytes. Para llamadas, cada llamada individual se redondea: incluso si la llamada duró solo un segundo, se contará como un minuto. Para tráfico web, las sesiones web individuales no se redondean. En vez de esto, el total del mes se redondea hacia arriba. Si alguien usa 1025 megabytes este mes, se le cobrarán 2 gigabytes.

La tabla users (datos sobre los usuarios):

* `user_id` — identificador único del usuario
* `first_name` — nombre del usuario
* `last_name` — apellido del usuario
* `age — edad` del usuario (en años)
* `reg_date` — fecha de suscripción (dd, mm, aa)
* `churn_date` — la fecha en que el usuario dejó de usar el servicio (si el valor es ausente, la tarifa se estaba usando cuando se recuperaron estos datos)
* `city` — ciudad de residencia del usuario
* `plan` — nombre de la tarifa

La tabla calls (datos sobre las llamadas):
* `id` — identificador único de la llamada
* `call_date` — fecha de la llamada
* `duration` — duración de la llamada (en minutos)
* `user_id` — el identificador del usuario que realiza la llamada

La tabla messages (datos sobre los SMS):

* `id` — identificador único del SMS
* `message_date` — fecha del SMS
* `user_id` — el identificador del usuario que manda el SMS

La tabla internet (datos sobre las sesiones web):
* `id` — identificador único de la sesión
* `mb_used` — el volumen de datos gastados durante la sesión (en megabytes)
* `session_date` — fecha de la sesión web
* `user_id` — identificador del usuario

La tabla plans (datos sobre las tarifas):
* `plan_name` — nombre de la tarifa
* `usd_monthly_fee` — pago mensual en dólares estadounidenses
* `minutes_included` — minutos incluidos al mes
* `messages_included` — SMS incluidos al mes
* `mb_per_month_included` — datos incluidos al mes (en megabytes)
* `usd_per_minute` — precio por minuto tras exceder los límites del paquete (por ejemplo, si el paquete incluye 100 minutos el operador cobrará el minuto 101)
* `usd_per_message` — precio por SMS tras exceder los límites del paquete
* `usd_per_gb` — precio por gigabyte de los datos extra tras exceder los límites del paquete (1 GB = 1024 megabytes)

#### Ingresos mensuales de cada usuario.

Para calcular esta cifra debemos hacer lo siguiente: restar el límite del paquete gratuito del número total de llamadas, SMS y datos; multiplicar el resultado por el valor de tarifa de llamadas; agregar el cargo mensual según la tarifa de llamadas.

```python
# Tabla de Consumo
consumo.head()
```

**Calcular ingresos por llamadas**

```python
# Desarrollamos una función que calcula el ingreso de los usuarios.
def total_income(row):

    # Límites de los paquetes.
    x = row['minutes_included']
    y = row['sms_included']
    z = row['gb_included']

    # Consumos de los usuarios.
    x1 = row['call_duration']
    y1 = row['total_sms']
    z1 = row['total_gb']
    x2 = row['no_calls']
    z2 = row['no_sessions']

    # Tarifas Adicionales.
    usd_calls = row['usd_per_minute']
    usd_inter = row['usd_per_gb']
    usd_plan  = row['usd_monthly_pay']
    usd_sms   = row['usd_per_message']

    # Calcular ingreso: Restar límite del paquete del no total de llamadas,sms y datos.
    # Multiplicar por el valor de la tarifa de llamadas.
    # Agregar cargo mensual según la tarfia de llamadas.

    exceso_mins = max(x1 - x, 0)
    exceso_msgs = max(y1 - y, 0)
    exceso_data = max(z1 - z, 0)
    income = exceso_mins * usd_calls + exceso_msgs * usd_sms + exceso_data * usd_inter + usd_plan

    return income
```

```python
consumo.apply(total_income,axis=1) # Probamos que funciona.
```

Creamos la nueva columna con total de ingresos.

```python
consumo['total_income'] = consumo.apply(total_income,axis=1) # Aplicamos función y creamos nueva columna.
consumo.head()                                               # Verificamos cambios.
```

```python
# Funcion para calcular el total de minutos consumidos al mes.
def total_mins(row):

    # Límites de los paquetes.
    x = row['minutes_included']
    y = row['sms_included']
    z = row['gb_included']

    # Consumos de los usuarios.
    x1 = row['call_duration']
    y1 = row['total_sms']
    z1 = row['total_gb']
    x2 = row['no_calls']
    z2 = row['no_sessions']

    # La diferencia entre el límite del plan y el consumo mensual del usuario.
    exceso_mins = max(x1 - x, 0)
    exceso_msgs = max(y1 - y, 0)
    exceso_data = max(z1 - z, 0)

    # El exceso más las llamadas realizadas al mes.
    return exceso_mins + x1
```

## Hipótesis

---

Hemos estudiado con éxito los parámetros más relevantes en nuestro punto anterior. Ahora en esta sección vamos a formular dos hipótesis interesantes que vamos a fallar en aceptar o rechazar según lo que nos indiquen las pruebas de nuestros datos.

* **Ho:** Los promedios de ingresos de los usuarios "Surf" y "Ultimate" son iguales
* **H1:** El ingreso promedio de los usuarios de las tarifas "Ultimate" y "Surf" difiere.

```python
# Hagamos un Filtro para los Usuarios "Surf" y otro para "Ultimate".
surf_filter = consumo[consumo['plan'] == 'surf']['total_income']
ultm_filter = consumo[consumo['plan'] == 'ultimate']['total_income']
```

Para probar nuestra hipótesis vamos a usar la prueba de la igualdad de las medias de dos poblaciones donde establecemos un nivel de significancia estadística (alpha) de 5%. Antes de aplicar nuestra prueba debemos saber si las varianzas son iguales o diferentes, vamos a revisar.

```python
surf_variance = np.var(surf_filter)
ultm_variance = np.var(ultm_filter)

print('La varianza de los usuarios "Surf" es:\n', surf_variance,)
print('-----------------------------------------')
print('La varianza de los usuarios "Ultimate" es:\n', ultm_variance)
```

Ahora sabemos que tenemos varianzas diferentes por lo que debemos establecer nuestro parámetro del test equal_var en false.

Vamos a nuestra prueba.

```python
surf_filter   # Dataset de usuarios surf
ultm_filter   # Dataset de usuarios ultimate

alpha  = 0.05 # Nivel de significancia estadística

result = st.ttest_ind(surf_filter, ultm_filter, equal_var=False)
print('p-value:', result.pvalue)

if result.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

La probabilidad de que nuestro valor estadístico (p-value) se encuentra por encima del valor de significancia (alpha), esto nos indica que rechazamos la hipótesis nula (**Ho**) y  nos declimanos por aceptar nuestra hipótesis alternativa (**H1**) que nos indica que los promedios de los ingresos de los usuarios surf y ultimate difieren.

**Segunda Hipótesis**

Se plantea que:

**H1:** El ingreso promedio de los usuarios en el área de estados Nueva York-Nueva Jersey es diferente al de los usuarios de otras regiones.

```python
# Hagamos un filtro de la ciudad y su ingreso.
city_filter  = consumo[consumo['city'] == 'New York-Newark-Jersey City, NY-NJ-PA MSA']['total_income']
other_cities = consumo[consumo['city'] != 'New York-Newark-Jersey City, NY-NJ-PA MSA']['total_income']
```

Para probar nuestra hipótesis vamos a usar la prueba de la igualdad de las medias de dos poblaciones donde establecemos un nivel de significancia estadística (alpha) de 5%. Antes de aplicar nuestra prueba debemos saber si las varianzas son iguales o diferentes, vamos a revisar.

```python
# Calculamos la varianza de nuestros filtros.
city_variance   = np.var(city_filter)
rest_cities_var = np.var(other_cities)

print('La varainza de los usuarios "Surf" es:\n', city_variance,)
print('-----------------------------------------')
print('La varainza de los usuarios "Ultimate" es:\n', rest_cities_var)
```

Nuestra probabilidad de que el valor calculado sea posible no supera el nivel de significancia estadística lo cual nos indica que nuestra hipótesis nula debe ser rechazada y que la hipótesis alternativa sobre que los ingresos de la ciudad de Nueva York-Nueva Jersey es diferente al de las demás ciudades es aceptada.
