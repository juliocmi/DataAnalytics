Houston tenemos un problema! La idea era imputar por moda pero resulta que nuestra moda son los valores ausentes. Entonces tendremos que buscar otra opción.

Nos tomamos la libertad de buscar en el sitio web oficial de ESRB alguna alternativa para nuestros valores ausentes y encontramos algo interesante.

En la guía de clasificación existen dos clasificaciones que no tienen una definición específica para videojuegos.

* `RP` - para videojuegos para todas las edades
* `RP+17` - para videojuegos que se piensa puede ser para mayores de 17 años.

Eureca, ahora vamos a usar estas clasificaciones para imputar nuestros valores por género según nuestro criterio de evaluación si el género corresponde o no a todas las edades o bien a mayores de 17 años. Puedes encontrar más información sobre estas clasificaciones en el apartado de fuentes.

Ahora vamos a sustituir primero nuestros valores ausentes por strings "Nan" para después reemplazarlos.

```python
games['rating'] = games['rating'].fillna('Nan')
games['rating'].unique()
```

vamos a desarrollar una función que nos ayude a reemplazar los ratings

```python
def replace_rating(data):
  """
  Funcion para Sustituir los Nan por Ratings de Videojuegos.
  """
  if data['genre'] == 'Action':
     x = data['rating'].replace('Nan', 'RP+17')
     return x
  if data['genre'] == 'Adventure':
     x1 = data['rating'].replace('Nan', 'RP')
     return x1
  if data['genre'] == 'Fighting':
     x2 = data['rating'].replace('Nan', 'RP+17')
     return x2
  if data['genre'] == 'Music':
     x3 = data['rating'].replace('Nan', 'RP')
     return x3
  if data['genre'] == 'Platform':
     x4 = data['rating'].replace('Nan', 'RP')
     return x4
  if data['genre'] == 'Puzzle':
     x5 = data['rating'].replace('Nan', 'RP')
     return x5
  if data['genre'] == 'Racing':
     x6 = data['rating'].replace('Nan', 'RP')
     return x6
  if data['genre'] == 'Role-Playing':
     x7 = data['rating'].replace('Nan', 'RP')
     return x7
  if data['genre'] == 'Shooter':
     x8 = data['rating'].replace('Nan', 'RP+17')
     return x8
  if data['genre'] == 'Simulation':
     x9 = data['rating'].replace('Nan', 'RP')
     return x9
  if data['genre'] == 'Sports':
     x10 = data['rating'].replace('Nan', 'RP')
     return x10
  if data['genre'] == 'Strategy':
     x11 = data['rating'].replace('Nan', 'RP+17')
     return x11
```

```python
# Aplicando función y verificando cambios
games['rating'] = games.apply(replace_rating, axis=1)
sorted(games['rating'].unique().tolist())
```

En este caso vamos a realizar una agrupación por plataforma y género tomando en cuenta de estas agrupaciones y de esta manera imputar los valores faltantes.

```python
games.groupby(['genre'])['critic_score'].median()
```

Vamos a desarrollar a una función que nos ayudará a reemplazar los valores ausentes utilizando las medianas de cada género.

```python
def replace_critics(data):
    """
    Función para reemplazar los valores ausentes en la columna "critic_score" agrupando por la mediana.
    """
    critic_ref = games.groupby(['genre'])['critic_score'].median()              # Agrupamos por valor de la mediana.

    data.loc[(data['genre'] == 'Action')       & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Action']
    data.loc[(data['genre'] == 'Adventure')    & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Adventure']
    data.loc[(data['genre'] == 'Fighting')     & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Fighting']
    data.loc[(data['genre'] == 'Music')        & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Music']
    data.loc[(data['genre'] == 'Platform')     & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Platform']
    data.loc[(data['genre'] == 'Puzzle')       & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Puzzle']
    data.loc[(data['genre'] == 'Racing')       & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Racing']
    data.loc[(data['genre'] == 'Role-Playing') & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Role-Playing']
    data.loc[(data['genre'] == 'Shooter')      & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Shooter']
    data.loc[(data['genre'] == 'Simulation')   & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Simulation']
    data.loc[(data['genre'] == 'Sports')       & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Sports']
    data.loc[(data['genre'] == 'Strategy')     & (data['critic_score'].isna()), 'critic_score'] = critic_ref.loc['Strategy']

    return data
```

```python
# aplicando la función y verificando
games = replace_critics(games)
games['critic_score'].unique()
```

En este punto desarrollaremos una función para reemplazar los valores ausentes en la variable *user_score* agrupando por la mediana.

```python
def replace_users(data):
  """
  # Función para reemplazar los valores ausentes en la columna "user_score" agrupando por la mediana.
  """
  user_ref = games.groupby(['genre'])['user_score'].median()  # Agrupamos por valor de la mediana.

  data.loc[(data['genre'] == 'Action')       & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Action']
  data.loc[(data['genre'] == 'Adventure')    & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Adventure']
  data.loc[(data['genre'] == 'Fighting')     & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Fighting']
  data.loc[(data['genre'] == 'Music')        & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Music']
  data.loc[(data['genre'] == 'Platform')     & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Platform']
  data.loc[(data['genre'] == 'Puzzle')       & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Puzzle']
  data.loc[(data['genre'] == 'Racing')       & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Racing']
  data.loc[(data['genre'] == 'Role-Playing') & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Role-Playing']
  data.loc[(data['genre'] == 'Shooter')      & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Shooter']
  data.loc[(data['genre'] == 'Simulation')   & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Simulation']
  data.loc[(data['genre'] == 'Sports')       & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Sports']
  data.loc[(data['genre'] == 'Strategy')     & (data['user_score'].isna()), 'user_score'] = user_ref.loc['Strategy']

  return data
```

```python
games = replace_users(games) # Aplicamos la función.
games['user_score'].unique() # Verificamos cambios.
```

Vamos a desarrollar una función de clasificación que se llamará *review_games* y nos ayudará a clasificar el puntaje de los videojuegos.

```python
def review_games(row): # Función para clasificar los videojuegos.
  if row['critic_score'] < 30:
    return 'Malo'
  elif row['critic_score'] < 50:
    return 'Regular'
  elif row['critic_score'] < 70:
    return 'Bueno'
  elif row['critic_score'] < 90:
    return 'Muy Bueno'
  else:
    return 'Excelente'
```

## Validar Hipótesis

Hemos estudiado con éxito los parámetros más relevantes en nuestro punto anterior. Ahora en esta sección vamos a formular dos hipótesis interesantes que vamos a fallar en aceptar o rechazar según lo que nos indiquen las pruebas de nuestros datos.

* **Ho:** Las calificaciones promedio para los usuarios de Xbox One y Pc son las mismas.

* **H1:** Las calificaciones promedio para los usuarios de Xbox One y Pc son diferentes.

```python
# Hagamos un Filtro para los Usuarios "Surf" y otro para "Ultimate".
x_one_filter = data_year[data_year['platform'] == 'XOne']['critic_score']
pc_filter    = data_year[data_year['platform'] == 'PC']['critic_score']
```

Para probar nuestra hipótesis vamos a usar la prueba de la igualdad de las medias de dos poblaciones donde establecemos un nivel de significancia estadística (alpha) de 5%. Antes de aplicar nuestra prueba debemos saber si las varianzas son iguales o diferentes, vamos a revisar.

```python
x_one_variance = np.var(x_one_filter)
pc_variance    = np.var(pc_filter)

print('La varianza de las calificaciones "XOne" es:\n', x_one_variance,)
print('-----------------------------------------')
print('La varianza de las calificaciones "PC" es:\n', pc_variance)
```

Ahora sabemos que tenemos varianzas diferentes el siguiente paso es realizar la prueba de hipótesis, para ello, debemos establecer nuestro parámetro del test equal_var en false.

```python
x_one_filter   # Dataset de calificaciones xbox
pc_filter      # Dataset de calificaciones pc

alpha  = 0.05 # Nivel de significancia estadística

result = st.ttest_ind(x_one_filter, pc_filter, equal_var=False)
print('p-value:', result.pvalue)

if result.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

La probabilidad de que nuestro valor estadístico (p-value) se encuentra por encima del valor de significancia (alpha), esto nos indica que rechazamos la hipótesis nula (**Ho**) y  nos declimanos por aceptar nuestra hipótesis alternativa (**H1**) que nos indica que los promedios de las calificaciones de las plataformas Xbox One y Pc son distintas.

**Segunda Hipótesis**

Se plantea que:

**Ho:** Las calificaciones promedio de los usuarios para los géneros de Accion y Deportes son Diferentes.

**H1:** Las calificaciones promedio de los usuarios para los géneros de Accion y Deportes son las Mismas.

```python
# Hagamos un Filtro para los Usuarios "Surf" y otro para "Ultimate".
action_filter = data_year[data_year['genre'] == 'Action']['user_score']
sports_filter = data_year[data_year['genre'] == 'Sports']['user_score']
```

Para probar nuestra hipótesis vamos a usar la prueba de la igualdad de las medias de dos poblaciones donde establecemos un nivel de significancia estadística (alpha) de 5%. Antes de aplicar nuestra prueba debemos saber si las varianzas son iguales o diferentes, vamos a revisar.

```python
action_variance = np.var(action_filter)
sports_variance = np.var(sports_filter)

print('La varianza de las calificaciones del género de Acción es:\n', action_variance,)
print('-----------------------------------------')
print('La varianza de las calificaciones del género de Deporte es:\n', sports_variance)
```

Tenemos varianzas diferentes por lo cual establecemos 'equal_var' en False.

Vamos a nuestra prueba.

```python
action_filter   # Dataset de calificaciones xbox
sports_filter      # Dataset de calificaciones pc

alpha  = 0.05 # Nivel de significancia estadística

result = st.ttest_ind(action_filter, sports_filter, equal_var=False)
print('p-value:', result.pvalue)

if result.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

Nuestra probabilidad de que el valor calculado sea posible no supera el nivel de significancia estadística lo cual nos indica que nuestra hipótesis nula debe ser rechazada y que la hipótesis alternativa sobre que las calificaciones de los usuarios para los géneros de acción y deportes son practicamente las mismas.
